```python
import torch
import torch.nn as nn
import torch.nn.functional as F

class SimpleNN(nn.Module):
    def __init__(self):
        super(SimpleNN,self).__init__()
        
        self.sc1=nn.Linear(784,128)
        self.sc2=nn.Linear(128,64)
        self.sc3=nn.Linear(64,10)
        
    def forward(self,x):
        x=F.relu(self.sc1(x))
        x=F.relu(self.sc2(x))
        x=F.log_softmax(self.sc3(x),dim=1)
        
        return x

model = SimpleNN()

print(model)

```

```python
# 1.Sigmoid
## 1) Logistic (or sigmoid) activation function is a non linear function that maps any input value between 0 and 1.  
## 2) sigmoid(x) = 1/(1 + exp(-x))

# 2.Tanh
## 1) Tanh is a non-linear activation function that outputs values between -1 and 1, with a mean output of 0. 
## 2) tanh(x) = (exp(x) - exp(-x)) / (exp(x) + exp(-x))


# 3.ReLU
## 1) (Rectified Linear Unit) is a non-linear activation function that outputs the input value if it is positive, or 0 if it is negative. computationally more efficient, well-suited for large-scale neural networks.

```

```python
import torch
import matplotlib.pyplot as plt

# Let's show sigmoid activation function!!
x = torch.linspace(-10, 10, 100)
y=torch.sigmoid(x)
plt.plot(x.numpy(), y.numpy(),color='purple')
plt.xlabel('x input')
plt.ylabel('y output')
plt.title("Sigmoid Activation Function")
plt.show()
```

```python
# Let's show ReLu then

y=torch.relu(x)

plt.plot(x.numpy(), y.numpy(),color='green')
plt.xlabel('x input')
plt.ylabel('y output')
plt.title("ReLU Activation Function")
plt.show()
```

```python
#Let's show tanh then!!

y=torch.tanh(x)


plt.plot(x.numpy(), y.numpy(),color='blue')
plt.xlabel('x input')
plt.ylabel('y output')
plt.title("Tanh Activation Function")
plt.show()
```

```python
class SimpleNN(nn.Module):
    def __init__(self, input_size, hidden_size, output_size):
        super(SimpleNN, self).__init__()
        self.fc1 = nn.Linear(input_size, hidden_size)  # 
        self.relu = nn.ReLU()                         # 
        self.fc2 = nn.Linear(hidden_size, output_size) # 

    def forward(self, x):
        x = x.view(-1, 28*28)  # 
        x = self.fc1(x)
        x = self.relu(x)
        x = self.fc2(x)
        return x

# 
input_size = 28 * 28
hidden_size = 128
output_size = 10
model = SimpleNN(input_size, hidden_size, output_size)
```
