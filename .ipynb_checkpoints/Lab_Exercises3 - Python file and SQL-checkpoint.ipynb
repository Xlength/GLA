```python
"""
Exercise 1 – Display the following menu to the user:
1)Create a new file
2)Display the file
3)Add a new item to the file
Make a selection to 1, 2 or 3
Ask the user to enter 1, 2 or 3. If they select anything other than 1, 2 and 3, 
it should display a suitable error message. 
If they select 1, ask the user to enter a school subject and save it to a new file called “Subjects.txt”. 
It should overwrite any existing file with a new file. 
If they select 2, display the contents of “Subjects.txt” file. 
If they select 3, ask the user to enter a new subject and save it to the file and then display the entire content of the file. 
Run the programme several times to test the options.
"""

print(' 1)Create a new file \n 2)Display the file \n 3)Add a new item to the file')
selection=int(input("Please enter 1, 2 or 3: "))

if selection==1:
    f = open("Subjects.txt", "w")
    subject=input("Please enter a school subject: ")
    f.write(subject)
    f.close()

elif selection==2:
    f = open("Subjects.txt", "r")
    print(f.read())
    f.close()

elif selection==3:
    f = open("Subjects.txt", "a")
    subject=input("Please enter a new school subject: ")
    f.write(subject)
    f.close()
    f = open("Subjects.txt", "r")
    print(f.read())
    f.close()

else:
    print("suitable error")
```

```python
'''
Exercise 3 - (Function and .csv files) Create the following menu:
1)Add to file
2)View all records
3)Quit programme
Enter the number of your selection:
If the user selects 1, allow them to add to a file called Salaries.csv which will store their name and salary.
If they select 2 it should display all records in the Salaries.csv file. 
If they select 3 it should stop the programme. 
If they select an incorrect option, they should see an error message. 
They should keep returning to the menu until they select option 3.
'''
import csv
print("1)Add to file")
print("2)View all records")
print("3)Quit programme")
selection=int(input("Enter the number of your selection:"))

while (True):
    if selection==1:
        f=open("Salaries.csv","a")
        name=input("Enter name:")
        salary=input("Enter salary:")
        f.write(name+': '+salary)
        f.close()
        print("1)Add to file")
        print("2)View all records")
        print("3)Quit programme")
        selection=int(input("Enter the number of your selection:"))

    elif selection==2:
        f=open("Salaries.csv","r")
        print(f.read())
        f.close()
        print("1)Add to file")
        print("2)View all records")
        print("3)Quit programme")
        selection=int(input("Enter the number of your selection:"))

    elif selection==3:
        print("Programme quit")
        f=open("Salaries.csv","w")
        f.close()
        break

    else:
        print("Incorrect option")
        print("1)Add to file")
        print("2)View all records")
        print("3)Quit programme")
        selection=int(input("Enter the number of your selection:"))
```

```python
'''
Exercise 4 – (function and csv) In Python, it is not technically possible to directly delete a record from a .csv file. 
Instead, you need to save the file to a temporary list in Python, 
make the changes to the list and then overwrite the original file with the temporary list.
Change the previous programme to allow you to do this. Your menu should now look like this:
1)Add to file
2)View all records
3)Delete a record
4)Quit programme
'''
# with open("data.csv", "r") as f:
#    reader = csv.reader(f)
#   data = list(reader)
# 这样会打印整个


import csv

def add_to_file():
    mess = input("Enter your message: ")
    with open("data.csv", "a", newline='') as f:
        writer = csv.writer(f)
        writer.writerow([mess])

def view_all_records():
    with open("data.csv", "r") as f:
        reader = csv.reader(f)
        for row in reader:
            print(', '.join(row))

def delete_record():
    record_to_delete = input("Enter the record you want to delete: ")
    
    # Read all records into a list
    records = []
    with open("data.csv", "r") as f:
        reader = csv.reader(f)
        for row in reader:
            records.append(row)
            print(records)
    # Filter out the record that matches the input

    records = [row for row in records if record_to_delete not in ''.join(row)]

    # Write the updated list back to the file
    with open("data.csv", "w") as f:
        for row in records:
            f.write(row)
    print(f"Record '{record_to_delete}' has been deleted (if it existed).")

while True:
    print("1) Add to file")
    print("2) View all records")
    print("3) Delete a record")
    print("4) Quit programme")
    input_mess = int(input("Enter your choice: "))
    
    if input_mess == 1:
        add_to_file()
    elif input_mess == 2:
        view_all_records()
    elif input_mess == 3:
        delete_record()
    elif input_mess == 4:
        print("Exiting program.")
        break
    else:
        print("Invalid choice, please try again.")
```
