E. **Arithmetic and ordering**
Compute the following results, storing the results in the variable specified and printing them out:

* `best_temp`, `best_hum` the best temperature and humidity to keep a snail for four weeks without it dying. *Look only at the four week exposures, ignoring snails kept for less than this time.* 

```python
deg_f = (data[:, 3] * 9/5) + 32 

mean_cols = np.mean(data, axis=0)


death_rates = data[:, 4] / data[:, 5]  
death_rate = np.sort(death_rates)  

exposure_death_order = data[np.argsort(death_rates), 1]  


four_week_exposures = data[data[:, 1] == 4]  
best_temp = four_week_exposures[np.argmin(four_week_exposures[:, 4] / four_week_exposures[:, 5]), 3]  
best_hum = four_week_exposures[np.argmin(four_week_exposures[:, 4] / four_week_exposures[:, 5]), 2]  

print(f"Temperatures in degrees Fahrenheit:\n{deg_f}")
print(f"Mean of all columns:\n{mean_cols}")
print(f"Sorted death rates:\n{death_rate}")
print(f"Exposure durations sorted by death rates:\n{exposure_death_order}")
print(f"Best temperature for 4-week exposures: {best_temp}°C")
print(f"Best humidity for 4-week exposures: {best_hum}%")
```

D) **Elementwise arithmetic** The image is stored as three colour planes, R,G,B. This is the `3` in the last position of the shape of the image (check the `shape` of the array where it is loaded in). The planes are often referred to as "channels". 

* Create an array `channel_diff`, which will be the **absolute** difference of the red channel of `img_array` (channel 0) from the green channel (channel 1), all scaled by a factor of **eight**. Hint: you need to use slicing here.


```python
import numpy as np
import matplotlib.pyplot as plt

img_array = ia.load_image_colour('imgs/parrots.png')

# 提取红色通道和绿色通道
red_channel = img_array[:, :, 0]  # 红色通道 (channel 0)
green_channel = img_array[:, :, 1]  # 绿色通道 (channel 1)

# 计算红色通道与绿色通道之间的绝对差异，并缩放
channel_diff = np.abs(red_channel - green_channel) * 8

# 显示结果
def show_image(img, title):
    plt.imshow(img, cmap='gray')  # 使用灰度图显示差异
    plt.axis('off')  # 不显示坐标轴
    plt.title(title)
    plt.show()

# 显示通道差异图像
show_image(channel_diff, 'Channel Difference (Red - Green)')

# 打印通道差异的形状和部分数据
print("Channel Difference shape:", channel_diff.shape)
print("Channel Difference data sample:", channel_diff[100:110, 100:110])  # 打印部分数据
```

E) **Stacking arrays**

Using *one* `for` loop, create a *list* of arrays which represent a panning animation which is a sequence of cropped versions of `img_array`. 

* The crops should start at [190,30] and be 200 pixels wide and 100 pixels high, and move 12 pixels right and 2 pixels up in each frame (that means the column index increases by 12, and the row index decreases by 2). 
* Create 30 frames. All of them will be the same size.
* Stack the array into a single array called `panning_array`.

Your code should be less than 15 lines (excluding any comments)!

```python
import numpy as np

img_array = ia.load_image_colour('imgs/parrots.png')

start_row, start_col = 190, 30
crop_width, crop_height = 200, 100
frames = []

# 创建 30 帧裁剪图像
for i in range(30):
    row = start_row - 2 * i  # 每帧向上移动 2 像素
    col = start_col + 12 * i  # 每帧向右移动 12 像素
    cropped = img_array[row:row + crop_height, col:col + crop_width]  # 裁剪图像
    frames.append(cropped)  # 添加到帧列表

# 将帧列表堆叠成一个数组
panning_array = np.stack(frames)

# 打印结果以确认
print("Panning array shape:", panning_array.shape)  # 应该是 (30, 100, 200, 3)
```

```python
ia.show_gif(panning_array, width="80%")
with tick.marks(10):
    assert(check_hash(panning_array,((30, 100, 200, 3), 834973352307.7242)))
```
