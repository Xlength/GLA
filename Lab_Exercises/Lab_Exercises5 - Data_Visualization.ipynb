```python
'''
Task3: Write a Python program to plot two or more lines with legends, different widths and colours.
'''
import matplotlib.pyplot as plt
import numpy as np

x=np.linspace(1,10,20)
y=np.sin(x)
y2=np.cos(x)
plt.figure(figsize=(15,3))
y3=x**2

plt.subplot(1,3,1)
plt.plot(x,y,'r--*',linewidth=3,label="y=sin(x)")
plt.title('Sin(x)')
plt.legend()
plt.subplot(1,3,2)
plt.plot(x,y2,'bo-',linewidth=4,label="y=cos(x)")
plt.title('Cos(x)')
plt.legend()
plt.subplot(1,3,3)
plt.plot(x,y3,'gd:',linewidth=5,label="y=x^2")
plt.title('x^2')
plt.legend()

plt.show()
```

```python
'''
Task 10: Write a Python programming to display a bar chart of the popularity of programming 
Languages. Attach a text label above each bar displaying its popularity (float value)
'''
# 编程语言及其流行度
languages = ['Python', 'Java', 'C', 'C++', 'R', 'JavaScript', 'C#']
popularity = [100, 96.3, 94.4, 87.5, 81.5, 79.4, 74.5]

# 设置颜色列表，确保每个条形图都有不同的颜色
colors = ['red', 'blue', 'green', 'orange', 'purple', 'cyan', 'magenta']

# 创建条形图
plt.bar(languages, popularity, color=colors)

# 为每个条形图添加流行度标签
for i in range(len(languages)):
    plt.text(i, popularity[i] + 1, f'{popularity[i]:.1f}', ha='center')

# 添加标题和标签
plt.title('Popularity of Programming Languages')
plt.xlabel('Programming Languages')
plt.ylabel('Popularity (%)')

# 显示图表
plt.show()
```

```python
'''
Task11: Write a Python programming to create a pie chart of the popularity of programming 
Languages
'''
explode=[0.1,0,0,0,0,0,0]
plt.pie(popularity,colors=colors, labels=languages,explode=explode, autopct='%1.1f%%')
plt.title('Popularity of Programming Languages')
plt.show()
```

```python
'''
Task 14: Write a Python program to create bar plots with error bars on the same figure.
Sample Data
Mean velocity: 0.2474, 0.1235, 0.1737, 0.1824
Standard deviation of velocity: 0.3314, 0.2278, 0.2836, 0.2645
'''

import matplotlib.pyplot as plt
import numpy as np

# 样本数据
mean_velocity = [0.2474, 0.1235, 0.1737, 0.1824]
std_dev_velocity = [0.3314, 0.2278, 0.2836, 0.2645]
labels = ['A', 'B', 'C', 'D']  # 为每个数据集添加标签

# 创建条形图，设置误差条
x = np.arange(len(mean_velocity))  # X轴位置
plt.bar(x, mean_velocity, yerr=std_dev_velocity, capsize=5, color='skyblue', alpha=0.7)

# 为每个条形图设置标签
plt.xticks(x, labels)

# 添加标题和标签
plt.title('Bar plot with Error Bars')
plt.xlabel('Samples')
plt.ylabel('Mean Velocity')

# 显示图表
plt.show()
```

```python
'''
Task 15: Write a Python program to create a stacked bar plot with error bars,
Use bottom to stack the women’s bars on top of the men’s bars.
Sample Data:
Means (men) = (22, 30, 35, 35, 26)
Means (women) = (25, 32, 30, 35, 29)
Men Standard deviation = (4, 3, 4, 1, 5)
Women Standard deviation = (3, 5, 2, 3, 3)
'''
import numpy as np
import matplotlib.pyplot as plt

# 样本数据
means_men = [22, 30, 35, 35, 26]
means_women = [25, 32, 30, 35, 29]
std_men = [4, 3, 4, 1, 5]
std_women = [3, 5, 2, 3, 3]

# 设置X轴的刻度标签
labels = ['A', 'B', 'C', 'D', 'E']
x = np.arange(len(labels))

# 绘制男性的条形图
plt.bar(x, means_men, yerr=std_men, capsize=5, label='Men', color='blue', alpha=0.7)

# 使用 'bottom' 参数，将女性的条形图堆叠在男性条形图上
plt.bar(x, means_women, yerr=std_women, capsize=5, bottom=means_men, label='Women', color='pink', alpha=0.7)

# 添加标题和标签
plt.xlabel('Group')
plt.ylabel('Scores')
plt.title('Stacked Bar Plot with Error Bars')

# 设置X轴的刻度和标签
plt.xticks(x, labels)

# 添加图例
plt.legend()

# 显示图表
plt.show()

```

```python
'''
Task 16: Write a Python program to draw a scatter graph taking a random distribution in X and Y and 
plotted against each other.
'''
import numpy as np
import matplotlib.pyplot as plt

# 生成随机分布的 X 和 Y 数据
np.random.seed(0)  # 设定随机种子，以确保每次运行时生成相同的随机数
x = np.random.rand(100)  # 生成 100 个介于 0 到 1 之间的随机数
y = np.random.rand(100)  # 生成 100 个介于 0 到 1 之间的随机数

# 创建散点图
plt.scatter(x, y, color='blue', alpha=0.5)

# 添加标题和标签
plt.title('Scatter Plot of Random Distribution')
plt.xlabel('X values')
plt.ylabel('Y values')

# 显示图表
plt.show()
```

```python
'''
Task 17: Write a Python program to draw a scatter plot using random distributions to generate balls 
of different sizes.
'''
import numpy as np
import matplotlib.pyplot as plt

# 生成随机分布的 X 和 Y 数据
np.random.seed(0)  # 确保每次运行时生成相同的随机数
x = np.random.rand(100)  # 生成 100 个随机 X 值
y = np.random.rand(100)  # 生成 100 个随机 Y 值

# 生成随机的点大小
sizes = np.random.rand(100) * 1000  # 生成 100 个随机点大小，乘以 1000 控制点的大小

# 创建散点图，设置点的大小
plt.scatter(x, y, s=sizes, color='blue', alpha=0.5)

# 添加标题和标签
plt.title('Scatter Plot with Balls of Different Sizes')
plt.xlabel('X values')
plt.ylabel('Y values')

# 显示图表
plt.show()
```
