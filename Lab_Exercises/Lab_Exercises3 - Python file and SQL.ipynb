```python
'''
Using the phonebook database, write a programme that will display the following menu
Main menu
1)View phone book
2)Add to phone book
3)Search for surname
4)Delete person from phone book
5)Quit
Enter your selection
If the user selects 1, they should be able to view the entire phonebook. 
If they select 2, it should allow them to add a new person to the phonebook. 
If they select 3, it should ask them for a surname and then display only the record of people with the same surname. 
If they select 4, it should ask for an ID and then delete that record from the table. 
If they select 5, it should end the programme. Finally, the programme should display a suitable message if they enter an incorrect selection from the menu. 
They should return to the menu after each action until they select 5.
'''
import sqlite3

def create_table():
    """Create the phonebook table if it doesn't exist."""
    conn = sqlite3.connect('phonebook.db')
    c = conn.cursor()
    c.execute('''
        CREATE TABLE IF NOT EXISTS phonebook (
            id INTEGER PRIMARY KEY AUTOINCREMENT,
            first_name TEXT NOT NULL,
            last_name TEXT NOT NULL,
            phone_number TEXT NOT NULL
        )
    ''')
    conn.commit()
    conn.close()

def view_phonebook():
    """View all records in the phonebook."""
    conn = sqlite3.connect('phonebook.db')
    c = conn.cursor()
    c.execute('SELECT * FROM phonebook')
    records = c.fetchall()
    conn.close()

    if records:
        print("\nPhonebook Records:")
        print("{:<5} {:<15} {:<15} {:<15}".format("ID", "First Name", "Last Name", "Phone Number"))
        print("-" * 55)
        for record in records:
            print("{:<5} {:<15} {:<15} {:<15}".format(record[0], record[1], record[2], record[3]))
    else:
        print("\nThe phonebook is empty.")

def add_to_phonebook():
    """Add a new person to the phonebook."""
    first_name = input("Enter first name: ").strip()
    last_name = input("Enter last name: ").strip()
    phone_number = input("Enter phone number: ").strip()

    conn = sqlite3.connect('phonebook.db')
    c = conn.cursor()
    c.execute('INSERT INTO phonebook (first_name, last_name, phone_number) VALUES (?, ?, ?)',
              (first_name, last_name, phone_number))
    conn.commit()
    conn.close()
    print(f"\n{first_name} {last_name} has been added to the phonebook.")

def search_by_surname():
    """Search for people by surname."""
    surname = input("Enter the surname to search for: ").strip()
    conn = sqlite3.connect('phonebook.db')
    c = conn.cursor()
    c.execute('SELECT * FROM phonebook WHERE last_name = ?', (surname,))
    records = c.fetchall()
    conn.close()

    if records:
        print(f"\nRecords matching surname '{surname}':")
        print("{:<5} {:<15} {:<15} {:<15}".format("ID", "First Name", "Last Name", "Phone Number"))
        print("-" * 55)
        for record in records:
            print("{:<5} {:<15} {:<15} {:<15}".format(record[0], record[1], record[2], record[3]))
    else:
        print(f"\nNo records found with surname '{surname}'.")

def delete_person():
    """Delete a person from the phonebook by ID."""
    try:
        person_id = int(input("Enter the ID of the person to delete: "))
    except ValueError:
        print("Invalid ID. Please enter a numeric value.")
        return

    conn = sqlite3.connect('phonebook.db')
    c = conn.cursor()
    c.execute('SELECT * FROM phonebook WHERE id = ?', (person_id,))
    record = c.fetchone()

    if record:
        c.execute('DELETE FROM phonebook WHERE id = ?', (person_id,))
        conn.commit()
        conn.close()
        print(f"\nRecord with ID {person_id} has been deleted.")
    else:
        conn.close()
        print(f"\nNo record found with ID {person_id}.")

def main_menu():
    """Display the main menu and handle user input."""
    create_table()  # Ensure the table exists

    while True:
        print("\nMain Menu")
        print("1) View phone book")
        print("2) Add to phone book")
        print("3) Search for surname")
        print("4) Delete person from phone book")
        print("5) Quit")
        try:
            selection = int(input("Enter your selection: "))
        except ValueError:
            print("Invalid input. Please enter a number between 1 and 5.")
            continue

        if selection == 1:
            view_phonebook()
        elif selection == 2:
            add_to_phonebook()
        elif selection == 3:
            search_by_surname()
        elif selection == 4:
            delete_person()
        elif selection == 5:
            print("Exiting the program.")
            break
        else:
            print("Invalid selection. Please choose a number between 1 and 5.")

if __name__ == "__main__":
    main_menu()
```
